# Student classifcation problem

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.ensemble as ens
import pandas as pd
from sklearn.model_selection import train_test_split

file  = "KaggleProblems\Student_performance_data.csv"

student_file = pd.read_csv(file)
```

First lets take a look at the data set by looking at the head and using df.describe()

```python
student_file.head()
```

```python
student_file.describe()
```

Now lets take a look to see any missing values and duplicates.

```python
student_file.drop_duplicates()
student_file.dropna()
```

```python
y = student_file['GradeClass']
X = student_file.drop(columns = "GradeClass")


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size  = 0.3)


X_test.shape
```

```python
from sklearn.metrics import accuracy_score, classification_report

model = ens.RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)


print(classification_report(y_test, pred))
```
